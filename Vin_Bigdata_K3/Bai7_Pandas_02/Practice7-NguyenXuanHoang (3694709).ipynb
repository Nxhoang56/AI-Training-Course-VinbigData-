### **Yêu cầu 2.3:**
---
Trong df_full: Tạo một cột Diem_he10 được tính dựa vào các cột tương ứng của từng hàng dữ liệu, theo công thức sau:

**Diem_he10 = 0.6 * A + 0.3 * ((B1+B2)/2) + 0.1 * ((C1+C2)/2)**
    
Làm tròn đến 1 số sau dấu phẩy

```python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY}
def score(A,B1, B2, C1, C2 ):
    Diem_he10 = 0.6 * A + 0.3 * ((B1+B2)/2) + 0.1 * ((C1+C2)/2)
    return round(Diem_he10,1)
df_full
```

```python
df_full['Diem_he10']=df_full.apply(lambda row: score(row['A'], row['B1'], row['B2'], row['C1'], row['C2'] ), axis = 1)
```

```python
df_full
```

### **Yêu cầu 2.4:**
---
Từ cột Diem_he10 trong df_full tạo một cột Diem_chu, Diem_so theo quy đổi:

    * Điểm hệ 10 Từ 9.0 đến 10.0  ----> Điểm chữ: A+  ----> Điểm số: 4.0
    * Điểm hệ 10 Từ 8.5 đến 8.9  ----> Điểm chữ: A    ----> Điểm số: 3.7
    * Điểm hệ 10 Từ 8.0 đến 8.4  ----> Điểm chữ: B+   ----> Điểm số: 3.5   
    * Điểm hệ 10 Từ 7.0 đến 7.9  ----> Điểm chữ: B    ----> Điểm số: 3.0
    * Điểm hệ 10 Từ 6.5 đến 6.9  ----> Điểm chữ: C+   ----> Điểm số: 2.5  
    * Điểm hệ 10 Từ 5.5 đến 6.4  ----> Điểm chữ: C    ----> Điểm số: 2.0     
    * Điểm hệ 10 Từ 5.0 đến 5.4  ----> Điểm chữ: D+   ----> Điểm số: 1.5 
    * Điểm hệ 10 Từ 4.0 đến 4.9  ----> Điểm chữ: D    ----> Điểm số: 1.0  
    * Điểm hệ 10 Từ 0.0 đến 3.9  ----> Điểm chữ: F    ----> Điểm số: 0.0 


```python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY}
def score2char(point):
    if point >= 9.0:
        return 'A+'
    elif 8.5 <= point <= 8.9:
        return 'A'
    elif 8.0 <= point <= 8.4:
        return 'B+'
    elif 7.0 <= point <= 7.9:
        return 'B'
    elif 6.5 <= point <= 6.9:
        return 'C+'
    elif 5.5 <= point <= 6.4:
        return 'C'
    elif 5.0 <= point <= 5.4:
        return 'D+'
    elif 4.0 <= point <= 4.9:
        return 'D'
    else:
        return 'F'
    
def score2four(point):
    if point >= 9.0:
        return 4.0
    elif 8.5 <= point <= 8.9:
        return 4.0
    elif 8.0 <= point <= 8.4:
        return 3.5
    elif 7.0 <= point <= 7.9:
        return 3.0
    elif 6.5 <= point <= 6.9:
        return 2.5
    elif 5.5 <= point <= 6.4:
        return 2.0
    elif 5.0 <= point <= 5.4:
        return 1.5
    elif 4.0 <= point <= 4.9:
        return 1.0
    else:
        return 0.0


```

```python
df_full['Diem_chu'] = df_full['Diem_he10'].apply(lambda x: score2char(x))

df_full['Diem_so'] = df_full['Diem_he10'].apply(lambda x: score2four(x))

df_full
```

### Xử lí dữ liệu khuyết thiếu

```python
print("Số lượng các giá trị missing trong tập dữ liệu")

print(df_patient.isnull().sum())
```

```python
def missing_values(df):
        mis_val = df.isnull().sum()
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Số giá trị Missing', 1 : 'Tỷ lệ % missing'})
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        'Tỷ lệ % missing', ascending=False).round(1)
        print ("File dữ liệu bao gồm có: " + str(df.shape[1]) + " cột.\n"      
            "Có " + str(mis_val_table_ren_columns.shape[0]) +
              " cột chứa missing values.")
        return mis_val_table_ren_columns

missing_values(df_patient)
```

```python
print("Những hàng thiếu dữ liệu:")

df_patient[df_patient.isnull().any(axis=1)]
```

```python
# Bỏ các hàng có trường Type bằng NaN

df_patient = df_patient[~df_patient['Type'].isnull()]

df_patient
```
