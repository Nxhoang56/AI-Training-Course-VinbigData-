**YÊU CẦU:** Vẽ biểu đồ giá gas của Japan, thiết lập các tham số để thu được biểu đồ như sau:
<img src='Pic/line1.jpg' width='800px'>

```python
#{HỌC VIÊN VIẾT CODE Ở ĐÂY}
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_gas_prices = pd.read_csv('Data_Excercise/gas_prices.csv')
df_gas_prices
```

```python
x = list(df_gas_prices['Year'])
y = list(df_gas_prices['Japan'])
```

```python
df_gas_prices['Japan'].describe()
```

```python
plt.figure(figsize = (10,3)) #Thiết lập kích thước biểu đồ

plt.plot(x,                 #Dữ liệu trục X
         y,                 #Dữ liệu trục Y
         color='g',       #Màu của đường
         linestyle='-.',            #Kiểu đường
         linewidth=1.0,            #Độ rộng của đường line
         marker='d',        #Kiểu điểm
         markersize = 8)           #Kích thước điểm

#Tiêu đề của đồ thị
plt.title('BIỂU ĐỒ THỂ HIỆN GIÁ GAS CỦA NHẬT BẢN GIAI ĐOẠN 1990-2008')
#Nhãn cho trục X
plt.xlabel('Thời gian')
#Nhãn cho trục Y
plt.ylabel('Giá GAS')

#Setup giới hạn cho trục X:
plt.xlim(1989,2009)

#Setup giới hạn cho trục Y:
plt.ylim(2.5,6)

#Setup tick cho trục X:
plt.xticks(x,
          rotation = 45)
#Hiển thị lưới:
plt.grid(ls = '--',
          axis = 'y')

plt.show()
```

**YÊU CẦU:** Trực quan hoá dữ liệu giá Gas của 4 nước: UK, USA, Australia, South Korea. với mục tiêu nhấn mạnh quốc gia có giá Gas cao nhất và thấp nhất. (như minh hoạ)

**YÊU CẦU:** Sử dụng dữ liệu đã có vẽ biểu đồ Pie chart và Bar chart như sau:
<img src='Pic/pie1.jpg' width='900px'>

```python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY}
#Tạo dữ liệu: Số lượng SV Nam - Nữ theo từng khoá.
label = ['K60','K61','K62','K63','K64','K65']
boys  = [200, 340, 260, 440, 300, 180]
girls = [30,  60,  90,  160, 180, 220]
```

```python
total = list(np.array(boys)+np.array(girls))
total
```

```python
sex = ['Nam', 'Nữ']
sum_boy = sum(boys)
sum_girl = sum(girls)
sum_std = [sum_boy, sum_girl]
```

```python
plt.figure(figsize = (12,5))
#Set up the frame size: 1*2 
#Fig1: piechart
plt.subplot(1,2,1)
plt.pie(sum_std,             #Giá trị thể hiện
        labels= sex,     #Nhãn tương ứng
        autopct='%.2f %%', #Tính toán và hiển thị % tương ứng
        pctdistance=0.7,   #Khoảng cách hiển thị giá trị % tới tâm.
        startangle=90,     #Góc bắt đầu của biểu đồ
        textprops={'color':'k','fontweight':'bold'}#thiết lập label
       ) #Label có xoay không?

plt.title('Biểu đồ sinh viên nam nữ khoa CNTT', fontdict={'fontname':'Arial',
                                                    'fontweight':'bold',
                                                   'fontsize':13})
plt.legend(loc=4)

#Fig2: bar chat
plt.subplot(1,2,2)
bar = plt.bar(label, 
        total,
        ) 
#Highlight the highest bar
bar[3].set_color('orange')
plt.title('Biểu đồ số lượng sv theo từng khóa', fontdict={'fontname':'Arial',
                                                    'fontweight':'bold',
                                                   'fontsize':13})
plt.grid(linestyle = '--', axis = 'y')

plt.show()
```

## THỰC HÀNH 4:
---
**YÊU CẦU:** Sử dụng dữ liệu đã có vẽ biểu đồ Pie chart và Line chart như sau:
<img src='Pic/pie2.jpg' width='900px'>

```python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY}
labels = ['K60','K61','K62','K63','K64','K65']
boys  = [200, 340, 260, 440, 300, 180]
girls = [30,  60,  90,  160, 180, 220]



```

```python
sv_total = list(np.array(boys)+np.array(girls))
sv_total
```

```python
plt.figure(figsize = (15,5)) #Thiết lập kích thước biểu đồ
#Vẽ biểu đồ đường trên khung 1:
plt.subplot(1,2,1) #Thiết lập Khung biểu đồ gồm 2 hàng 1 cột

plt.plot(labels, sv_total,'r-s',lw=1.0, ms=7, label = 'Tổng số SV')  #Vẽ biểu đồ đường plot 1
plt.plot(labels, boys,'b--o',lw=1.0, ms=7, label = 'Sinh vien Nam')  #Vẽ biểu đồ đường plot 2
plt.plot(labels, girls,'k--o',lw=1.0, ms=7, label = 'Sinh vien Nữ')  #Vẽ biểu đồ đường plot 3


plt.title('Biểu đồ số lượng sv nam nữ và tổng theo khóa', fontdict={'fontname':'Arial',
                                                                    'fontweight':'bold',
                                                                    'fontsize':13})
plt.legend(loc=4)
plt.grid(ls='--')

#----------------------------------------------------------------
#Vẽ biểu đồ Bar trên khung 2:
plt.subplot(1,2,2)

e = [0.3,0,0.1,0.2,0.1,0.2]
plt.pie(sv_total, 
        labels=labels,
        textprops={'color':'k','fontweight':'bold'},
        rotatelabels=False,
        labeldistance=1.05,
        wedgeprops=dict(width=0.6,edgecolor='w'), #Xác định độ rộng của donus
        autopct='%.2f %%', 
        pctdistance=0.6,
         explode = e)

plt.title('Tỷ lệ sinh viên của từng khóa', fontdict={'fontname':'Tahoma',
                                                  'fontweight':'bold',
                                                   'fontsize':18}
         )

plt.show()
```

```python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY}

df_diamonds = pd.read_csv('Data_Excercise/Diamonds.txt',
                         sep = '\t',
                         names = ['Weight(cara)','Price(USD)'],
                         header = None)
df_diamonds.head(3)
```

```python
# weight = list(df_diamonds['Weight(cara)'].values)
# price = list(df_diamonds['Price(USD)'].values)
```

```python
plt.figure(figsize = (10,10))
plt.subplot(2,1,1)
#Vẽ biểu đồ hộp Boxplot:
plt.boxplot(df_diamonds['Weight(cara)'],           #Dữ liệu đầu vào
            widths=[0.5],           #Độ rộng của Box
            labels=['Weight'],
           vert = False, 
           showmeans = True) #Nhãn của Boxplot

plt.grid( axis = 'x', ls='--')
plt.xlabel('Trọng lượng')
plt.xticks([0,1,2,3,4,5])

plt.subplot(2,1,2)

plt.boxplot(df_diamonds['Price(USD)'],           #Dữ liệu đầu vào
            widths=[0.5],           #Độ rộng của Box
            labels=['Prices'],#Nhãn của Boxplot
           vert = False,
           showmeans = True) #show mean value 

plt.grid( axis = 'x', ls='--')
plt.xlabel('Giá')
plt.xticks()
plt.show()
```

```python
#Gọi các thư viện sử dụng
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#---------------------------------
import warnings
warnings.filterwarnings('ignore')
```

```python
#Đọc dữ liệu:
data500 = pd.read_csv('Data_Excercise/Data_500.csv', 
                      sep=';')

data500.head()
```

# Yêu cầu 1:
---
Thực hiện thống kê số lượng Nam – Nữ trong tập dữ liệu và trực quan hóa kết quả:

```python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY}

data500.describe()
```

```python
male = data500[data500['Gender'] == 'Male'].shape[0]
male
```

```python
female = data500[data500['Gender'] == 'Female'].shape[0]
female
```

```python
plt.figure(figsize = (12,5))
#Set up the frame size: 1*2 
#Fig1: piechart
plt.subplot(121)
plt.pie([male, female],             #Giá trị thể hiện
        labels= ['Nam', 'Nữ'],     #Nhãn tương ứng
        autopct='%.2f %%', #Tính toán và hiển thị % tương ứng
        pctdistance=0.7,   #Khoảng cách hiển thị giá trị % tới tâm.
        startangle=90,     #Góc bắt đầu của biểu đồ
        textprops={'color':'k','fontweight':'bold'}#thiết lập label
       ) #Label có xoay không?

plt.title('Biểu đồ số phần trăm sinh viên nam nữ', fontdict={'fontname':'Arial',
                                                    'fontweight':'bold',
                                                   'fontsize':13})
plt.legend(loc=4)

#Fig2: bar chat
plt.subplot(122)
bar = plt.bar(['Nam', 'Nữ'], [male, female]) 
#Highlight the highest bar
plt.title('Biểu đồ thống kê số lượng sinh viên nam nữ', fontdict={'fontname':'Arial',
                                                    'fontweight':'bold',
                                                   'fontsize':13})
plt.grid(linestyle = '--', axis = 'y')

plt.show()
```
