```python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY}
df_temp = pd.read_csv("Data_Excercise\Data_Temperature.csv")

df_temp.info()


```

```python
df_temp.head()
```

```python
# Kiểm tra dữ liệu missing 
df_temp.isnull().sum()
```

```python
# Liệt kê các thời điểm không có dữ liệu
df_temp[df_temp.isnull().any(axis=1)]
```

```python
# Chuyển TimeVN về dạng chuỗi thời gian
df_temp['TimeVN'] = pd.to_datetime(df_temp['TimeVN'])

df_temp['TimeVN'].info()
```

```python
# Thiết lập cột TimeVN làm index

df_temp.set_index('TimeVN', inplace=True)
df_temp
```

### Khai phá dữ liệu

```python
print("Nhiệt độ trung bình của Hà Giang là:", df_temp['HaGiang'].mean())
print("Độ lệch chuẩn nhiệt độ của Hà Giang là:", df_temp['HaGiang'].std())
```

```python
print("Nhiệt độ trung bình của Cà Mau là:", df_temp['CaMau'].mean())
print("Độ lệch chuẩn nhiệt độ của Cà Mau là:", df_temp['CaMau'].std())
```

#### Đặc trưng chuỗi thời gian

#### 1) Xu hướng nhiệt độ trung bình qua từng năm

```python
yearly = df_temp.resample('Y').mean()

yearly
```

```python
yearly.plot(style=[':*', '--o'], color=['red','green'])

plt.ylabel('Nhiệt độ trung bình')
plt.xlabel('Năm')
plt.title('ĐỒ THỊ THỐNG KÊ NHIỆT ĐỘ TRUNG BÌNH THEO CÁC NĂM')
```

```python
# Từ biểu đồ nhiệt đồ ta thấy nhiệt độ các năm ở Cà Mau đều cao hơn so với Hà Giang với độ chênh lệch xấp xỉ 4o
```

#### 2) Thống kê nhiệt độ theo mùa vụ

```python
seasonly = df_temp.resample('M').mean()

seasonly
```

```python
seasonly.plot(style=['-', '-'], color=['red','green'])
plt.ylabel('Nhiệt độ trung bình theo tháng');
plt.show()
```

```python
# Hiển thị biểu đồ nhiệt độ trong năm 2018 ở Hà Giang và Cà Mau
df_year_2018 = df_temp.loc['2018']

monthly = df_year_2018.resample("M").mean()

monthly.plot(style=['-', '-'], color=['red','green'])
plt.ylabel('Nhiệt độ trung bình theo tháng trong năm 2018');
plt.show()
```

```python
# Nhận xét:
# + Nhiệt độ trung bình trong năm của cả hai tỉnh đều có tính chất mùa vụ
# + Nhiệt độ của Cà Mau khá là ổn định và nó dao động trong khoảng 27o
# + Trong khi ở Hà Giang có sự biến động lớn, nóng từ tháng 5 đến tháng 10 trong khi những tháng còn lại trời lạnh
```

#### 3) Nhiệt độ trung bình trong ngày

```python
by_time = df_temp.groupby(df_temp.index.time).mean()

by_time
```

```python
#Trực quan hóa dữ liệu:
import numpy as np

hourly_ticks = 3 * 60 * 60 * np.arange(8) + 60 * 60
by_time.plot(xticks=hourly_ticks, style=[':*', '--s'])
plt.grid()
plt.xlabel('Thời gian (h)')
plt.ylabel('Nhiệt độ trung bình theo giờ')
plt.title('BIỂU ĐỒ THỂ HIỆN NHIỆT ĐỘ TRUNG BÌNH TRONG NGÀY CỦA HAI TỈNH THÀNH',loc='center')
plt.show()
```

```python
# Nhận xét: Xu hướng tăng giảm nhiệt độ trong ngày của hai tỉnh thành tương đối giống nhau tuy nhiên nhiệt độ
# tại Cà Mau luôn cao hơn.
```
