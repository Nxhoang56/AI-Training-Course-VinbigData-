```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#iseas.edu.sg/articles-commentaries/iseas-perspective/2021-28-vietnams-solar-power-boom-policy-implications-for-other-asean-member-states-by-thang-nam-do-and-paul-j-burke/#:~:text=Vietnam%20has%20recently%20seen%20a,in%202019%20(Figure%201).
year = [2017,2018,2019]
solar_consumption = pd.Series([10, 86, 4750], index = ['2017', '2018', '2019'])
vn_gdp = pd.Series([7.15, 7.2, 6.94], index = ['2017', '2018', '2019'])
```

```python
plt.scatter(solar_consumption.values, vn_gdp.values)
plt.xlabel('Solar consumption')
plt.ylabel('Vietnam GDP')
plt.show()
```

```python
plt.figure(figsize = (12,6))

plt.grid()
#solar consumption - bar chart
plt.bar(year, solar_consumption.values, color = 'g')
plt.ylabel('Solar installed capacity (MW)',
           color = 'green',
           fontdict = {'fontsize':14})

plt.twinx() 

#GDP line chart
plt.plot(year, vn_gdp.values, 'b-o', lw = 2)

plt.ylabel('GDP %',
           color = 'blue',
           fontdict = {'fontsize':14})

# plt.title('BIỂU ĐỒ THỂ HIỆN DOANH SỐ VÀ LỢI NHẬN CỦA CÁC NĂM',
#          fontdict = {'fontsize':18})

plt.xticks(year)
plt.show()

# https://www.statista.com/statistics/1006138/vietnam-total-solar-energy-capacity/
```
