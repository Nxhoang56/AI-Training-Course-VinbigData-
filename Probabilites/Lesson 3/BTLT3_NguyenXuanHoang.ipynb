```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

import warnings
warnings.filterwarnings("ignore")
```

<h2>Bài tập 3.1:</h2>
<ol>
<li> Tạo ra 1000 giá trị của biến ngẫu nhiên tuân theo luật phân phối
nhị thức với tham số n = 30 và p = 0.6. </li>
<li> Tìm xác suất để biến ngẫu nhiên nhận giá trị là 20. </li>
<li> Tìm xác suất để biến ngẫu nhiên có giá trị nhỏ hơn hoặc bằng 18. </li>
<li> Tìm kỳ vọng, phương sai, độ lệch tiêu chuẩn, trung vị. </li>
<ol>

```python
x = np.random.binomial(30, 0.6, 1000)

plt.hist(x)
plt.xlim(0, 30)
plt.show()
```

```python
X = stats.binom(60, 0.3)

print("Xác suất để biến ngẫu nhiên nhận giá trị bằng 20 là:", X.pmf(20))
print("Xác suất để biến ngẫu nhiên có giá trị nhỏ hơn hoặc bằng 18 là:", np.mean([X.pmf(i) for i in range(19)]))
```

```python
print("Kì vọng là:", X.mean())
print("Phương sai là:", X.var())
print("Độ lệch chuẩn là:", X.std())
print("Trung vị là:", X.median())
```

<h2>Bài tập 3.2:</h2>
<ol>
<li> Tạo ra n= 500 số ngẫu nhiên tuân theo luật phân phối Fisher với
bậc tự do là 6 và 8. </li>
<li> Tìm kỳ vọng, phương sai, độ lệch tiêu chuẩn </li>
<li> Tìm giá trị phân vị mức 0.9; 0.95; 0.99 </li>
<li> Vẽ đồ thị của hàm mật độ </li>
<li> Vẽ đồ thị hàm phân phối </li>
<ol>

```python
x = np.random.f(6, 8, 500)
x
```

```python
F = stats.f(6, 8)

print("Kì vọng:", F.mean())
print("Phương sai:", F.var())
print("Độ lệch chuẩn:", F.std())
```

```python
print("Phân vị 0.9:", np.quantile(x, 0.9))
print("Phân vị 0.95:", np.quantile(x, 0.95))
print("Phân vị 0.99:", np.quantile(x, 0.99))
```

```python
a = np.linspace(0, 20, 1000)
p = F.pdf(a)

plt.plot(a, p)
plt.title("ĐỒ THỊ HÀM MẬT ĐỘ")
plt.show()
```

```python
a = np.linspace(0, 20, 1000)
p = F.cdf(a)

plt.plot(a, p, c='r')
plt.title("ĐỒ THỊ HÀM PHÂN PHỐI")
plt.show()
```
