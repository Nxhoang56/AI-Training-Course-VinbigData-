# Bài tập 3.1


```python
import numpy as np 
import scipy.stats as stats 

# Tạo ra 5000 biến ngẫu nhiên tuân theo quy luật phân phối Poisson với Lamda = 30
X = stats.poisson.rvs(30, size=5000)

# Tìm xác suất để biến ngẫu nhiên nhận giá trị là 35
p1 = X[X == 35].size/X.size
print("Xác suất để biến ngẫu nhiên nhận giá trị là 35:", p1)

# Tìm xác suất để biến ngẫu nhiên nhận giá trị nhỏ hơn hoặc bằng 15 
p2 = X[(X <= 15)].size/X.size 
print("\nXác suất để biến ngẫu nhiên nhận giá trị nhỏ hơn hoặc bằng 15:", p2)

# Tìm kỳ vọng, phương sai, độ lệch tiêu chuẩn, trung vị
print("\nKỳ vọng:", stats.poisson.mean(30))
print("Phương sai:", stats.poisson.var(30))
print("Độ lệch chuẩn:", stats.poisson.std(30))
print("Trung vị:", stats.poisson.median(30))
```

# Bài tập 3.2

```python
import numpy as np 
import scipy.stats as stats 
import matplotlib.pyplot as plt 
 
# Tạo ra 1000 số ngẫu nhiên tuân theo luật phân phối Student với bậc tự do 15
X = stats.t.rvs(15, size=1000)

# Tìm kỳ vọng, phương sai, độ lệch tiêu chuẩn, trung vị
print("\nKỳ vọng:", stats.t.mean(15))
print("Phương sai:", stats.t.var(15))
print("Độ lệch chuẩn:", stats.t.std(15))

# Tìm giá trị phân vị mức 0.9, 0.95, 0.99
p_09 = stats.t.ppf(0.9, 15)
p_095 = stats.t.ppf(0.95, 15)
p_099 = stats.t.ppf(0.99, 15)
print("\nPhân vị mức 0.9:", p_09)
print("Phân vị mức 0.95:", p_095)
print("Phân vị mức 0.99:", p_099)

# Vẽ đồ thị của hàm mật độ
x = np.arange(-5,5, 0.001)
plt.plot(x, stats.t.pdf(x, 15), color="red")
plt.show()

# Vẽ đồ thị của hàm phân phối
```
