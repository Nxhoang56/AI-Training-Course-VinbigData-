```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Bài tập 5.1:
Để giúp xác định nhu cầu về các khoá học đánh golf, một cuộc khảo sát đã 
được thực hiện. Một mẫu gồm 75 người chơi golf được hỏi họ đã chơi bao nhiêu lần 
trong năm ngoái. Dữ liệu thu được như sau. 
1. Tính các giá trị thống kê min, max, Q1, Q2, Q3 
2. Vẽ đồ thị boxplot để mô tả các giá trị đặc trưng 
3. Vẽ đồ thị tần số, tần xuất, hàm mật độ có histogram, chỉ có hàm mật độ.

```python
X = pd.read_csv("Lesson5_dataBTVN1.csv")
X = X.values.reshape(-1)

print(X)
```

```python
print("Min:", X.min())
print("Max:", X.max())
print("Q1:", np.quantile(X, 0.25))
print("Q2:", np.quantile(X, 0.5))
print("Q3:", np.quantile(X, 0.75))
```

```python
plt.boxplot(X)
plt.title("ĐỒ THỊ BOXPLOT CỦA TẬP DỮ LIỆU X")
plt.show()
```

```python
plt.hist(X)
plt.title("ĐỒ THỊ TẦN SỐ")
plt.show()
```

```python
plt.hist(X, density=True)
plt.title("ĐỒ THỊ TẦN SUẤT")
plt.show()
```

```python
x, pmf = np.unique(X, return_counts=True)
pmf = pmf / len(X)

plt.hist(X, density=True)
plt.plot(x, pmf)

plt.show()
```

```python
plt.plot(x, pmf)
plt.title("PMF")
plt.show()
```

### Bài tập 5.2
1. Tạo 500 số ngẫu nhiên (X, Y) có phân phối chuẩn 2 chiều với mean = [100, 50], 
cov=[[36, 0.3], [0.2, 9]] 
2. Tính các giá trị thống kê min, max, Q1, Q2, Q3 của X, Y 
3. Vẽ đồ thị boxplot để mô tả các giá trị đặc trưng của X, Y 
4. Vẽ đồ thị tần số, tần xuất, hàm mật độ có histogram, chỉ có hàm mật độ đồng thời 
của (X, Y)

```python
mean = np.array([100, 50])
cov = [[100, 0.4], [0.4, 36]]

data = np.random.multivariate_normal(mean, cov, 5000)
```

```python
X, Y = data[:, 0], data[:, 1]
```

```python
print("Thống kê theo X.")
print("Min:", X.min())
print("Max:", X.max())
print("Q1:", np.quantile(X, 0.25))
print("Q2:", np.quantile(X, 0.5))
print("Q3:", np.quantile(X, 0.75))
```

```python
print("Thống kê theo Y.")
print("Min:", Y.min())
print("Max:", Y.max())
print("Q1:", np.quantile(Y, 0.25))
print("Q2:", np.quantile(Y, 0.5))
print("Q3:", np.quantile(Y, 0.75))
```

```python
plt.boxplot(X)
plt.title("ĐỒ THỊ BOXPLOT THEO X")
plt.show()
```

```python
plt.boxplot(Y)
plt.title("ĐỒ THỊ BOXPLOT THEO Y")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
print("Đồ thị theo X")
X = np.round(X, 0)

plt.subplot(2, 2, 1)
plt.hist(X)
plt.title("ĐỒ THỊ TẦN SỐ")

plt.subplot(2, 2, 2)
plt.hist(X, density=True)
plt.title("ĐỒ THỊ TẦN SUẤT")

plt.subplot(2, 2, 3)
x, pmf = np.unique(X, return_counts=True)
pmf = pmf / len(X)

plt.hist(X, density=True)
plt.plot(x, pmf)

plt.subplot(2, 2, 4)
plt.plot(x, pmf)
plt.title("ĐỒ THỊ HÀM MẬT ĐỘ")

plt.show()
```

```python
plt.figure(figsize=(10, 6))
print("Đồ thị theo Y")
Y = np.round(Y, 0)

plt.subplot(2, 2, 1)
plt.hist(Y)
plt.title("ĐỒ THỊ TẦN SỐ")

plt.subplot(2, 2, 2)
plt.hist(Y, density=True)
plt.title("ĐỒ THỊ TẦN SUẤT")

plt.subplot(2, 2, 3)
x, pmf = np.unique(Y, return_counts=True)
pmf = pmf / len(Y)

plt.hist(Y, density=True)
plt.plot(x, pmf)

plt.subplot(2, 2, 4)
plt.plot(x, pmf)
plt.title("ĐỒ THỊ HÀM MẬT ĐỘ")

plt.show()
```
