```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## Bài thực hành 1

```python
# Tạo 5000 số ngẫu nhiên có phân phối nhị thức với n=50, p=0.7

X = np.random.binomial(50,0.7,size=5000)

# Tính các giá trị thống kê min, max, Q1, Q2, Q3
print("Giá trị min:",np.min(X))
print("Giá trị max:",np.max(X))
print("Giá trị Q1:",np.quantile(X,0.25))
print("Giá trị Q2:",np.quantile(X,0.5))
print("Giá trị Q3:",np.quantile(X,0.75))

# Vẽ đồ thị boxplot để mô tả các giá trị đặc trưng
plt.boxplot(X)
plt.show()

# Vẽ đồ thị tần số, tần xuất, hàm mật độ có histogram, chỉ có hàm mật độ
```

```python
# Tần số

# plt.figure(figsize=(15,5))
plt.hist(X, bins=10)
plt.title("Đồ thị tần số")
plt.xlabel('X')
plt.ylabel('Tần số')
plt.show()
```

```python
# Tần suất

# plt.figure(figsize=(15,5))
plt.hist(X, bins=10, density=True)
plt.title("Đồ thị tần suất")
plt.xlabel('X')
plt.ylabel('Tần suất')
plt.show()
```

```python
# Hàm mật độ có histogram

# plt.figure(figsize=(10,6))
kwargs = dict(hist_kws={'alpha':0.5}, kde_kws={"linewidth":2})
sns.distplot(X, color='green',bins=10,label="Hàm mật độ")
plt.legend(loc="upper left")
plt.show()
```

```python
# Hàm mật độ không có histogram

plt.figure(figsize=(10,6))
kwargs = dict(hist_kws={'alpha':0.5}, kde_kws={"linewidth":2})
sns.distplot(X, color='green',bins=10,hist=False, label="Hàm mật độ")
plt.legend(loc="upper left")
plt.show()
```

## Bài thực hành 2

```python
# Tạo 5000 số ngẫu nhiên (Chiều cao, cân nặng) có phân phối chuẩn 2 chiều với mean=[165, 55], cov=[[100, 0.4], [0.4, 36]]
mean=[165, 55]
cov=[[100, 0.4], [0.4, 36]]
X, Y = np.random.multivariate_normal(mean, cov, size=5000).T

# Tính các giá trị thống kê min, max, Q1, Q2, Q3 của chiều cao, cân nặng
print("Các giá trị thống kê của chiều cao:")
print("Min: ", np.min(X))
print("Max: ", np.max(X))
print("Q1: ", np.quantile(X, 0.25))
print("Q2: ", np.quantile(X, 0.5))
print("Q3: ", np.quantile(X, 0.75))

print("\nCác giá trị thống kê của cân nặng:")
print("Min: ", np.min(Y))
print("Max: ", np.max(Y))
print("Q1: ", np.quantile(Y, 0.25))
print("Q2: ", np.quantile(Y, 0.5))
print("Q3: ", np.quantile(Y, 0.75))
```

```python
# Vẽ đồ thị boxplot để mô tả các giá trị đặc trưng của chiều cao, cân nặng

plt.figure(figsize=(15,6))

plt.subplot(1,2,1)
plt.boxplot(X)
plt.title("Chiều cao", color='r')
plt

plt.subplot(1,2,2)
plt.boxplot(Y)
plt.title("Cân nặng", color='r')

plt.show()
```

```python
# Vẽ đồ thị tần số, tần xuất, hàm mật độ có histogram, chỉ có hàm mật độ đồng thời của chiều cao, cân nặng

plt.figure(figsize=(15,25))

# Tần số chiều cao
plt.subplot(4,2,1)
plt.hist(X, bins=10,color='b')
plt.title("Đồ thị tần số của chiều cao")
plt.xlabel('Chiều cao (cm)')
plt.ylabel('Tần số')

# Tần số cân nặng
plt.subplot(4,2,2)
plt.hist(Y, bins=10, color='g')
plt.title("Đồ thị tần số của cân nặng")
plt.xlabel('Cân nặng (kg)')
plt.ylabel('Tần số')


# Tần suất chiều cao
plt.subplot(4,2,3)
plt.hist(X, bins=10, density=True,color='blue')
plt.title("Đồ thị tần suất của chiều cao")
plt.xlabel('Chiều cao (cm)')
plt.ylabel('Tần suất')

# Tần suất cân nặng
plt.subplot(4,2,4)
plt.hist(Y, bins=10, density=True, color='green')
plt.title("Đồ thị tần suất của cân nặng")
plt.xlabel('Cân nặng (kg)')
plt.ylabel('Tần suất')

# Hàm mật độ có histogram của chiều cao
plt.subplot(4,2,5)
kwargs = dict(hist_kws={'alpha':0.5}, kde_kws={"linewidth":2})
sns.distplot(X, color='blue',bins=10,label="Hàm mật độ")
plt.legend(loc="upper left")
plt.title("Hàm mật độ có histogram của chiều cao")


# Hàm mật độ có histogram của cân nặng
plt.subplot(4,2,6)
kwargs = dict(hist_kws={'alpha':0.5}, kde_kws={"linewidth":2})
sns.distplot(Y, color='green',bins=10,label="Hàm mật độ")
plt.legend(loc="upper left")
plt.title("Hàm mật độ có histogram của cân nặng")

# Hàm mật độ không có histogram của chiều cao
plt.subplot(4,2,7)
kwargs = dict(hist_kws={'alpha':0.5}, kde_kws={"linewidth":2})
sns.distplot(X, color='blue',bins=10,hist=False,label="Hàm mật độ")
plt.legend(loc="upper left")
plt.title("Hàm mật độ không có histogram của chiều cao")


# Hàm mật độ không có histogram của cân nặng
plt.subplot(4,2,8)
kwargs = dict(hist_kws={'alpha':0.5}, kde_kws={"linewidth":2})
sns.distplot(Y, color='green',bins=10,hist=False,label="Hàm mật độ")
plt.legend(loc="upper left")
plt.title("Hàm mật độ không có histogram của cân nặng")

plt.show()
```
