```python
import numpy as np
import matplotlib.pyplot as plt
```

### Bài tập thực hành 5.1
1. Tạo 5000 số ngẫu nhiên có phân phối nhị thức với n=50, p=0.7
2. Tính các giá trị thống kê min, max, Q1, Q2, Q3
3. Vẽ đồ thị boxplot để mô tả các giá trị đặc trưng
4. Vẽ đồ thị tần số, tần xuất, hàm mật độ có histogram, chỉ có hàm
mật độ.

```python
X = np.random.binomial(50, 0.7, 5000)
```

```python
print("Min:", X.min())
print("Max:", X.max())
print("Q1:", np.quantile(X, 0.25))
print("Q2:", np.quantile(X, 0.5))
print("Q3:", np.quantile(X, 0.75))
```

```python
plt.boxplot(X)
plt.title("ĐỒ THỊ BOXPLOT CỦA TẬP DỮ LIỆU X")
plt.show()
```

```python
plt.hist(X)
plt.title("ĐỒ THỊ TẦN SỐ")
plt.show()
```

```python
plt.hist(X, density=True)
plt.title("ĐỒ THỊ TẦN SUẤT")
plt.show()
```

```python
x, pmf = np.unique(X, return_counts=True)
pmf = pmf / len(X)

plt.hist(X, density=True)
plt.plot(x, pmf)

plt.show()
```

```python
plt.plot(x, pmf)
plt.show()
```

### Bài thực hành số 5.2
1. Tạo 5000 số ngẫu nhiên (Chiều cao, cân nặng) có phân phối chuẩn
2 chiều với mean=[165, 55], cov=[[100, 0.4], [0.4, 36]]
2. Tính các giá trị thống kê min, max, Q1, Q2, Q3 của chiều cao, cân
nặng
3. Vẽ đồ thị boxplot để mô tả các giá trị đặc trưng của chiều cao, cân
nặng
4. Vẽ đồ thị tần số, tần xuất, hàm mật độ có histogram, chỉ có hàm
mật độ đồng thời của chiều cao, cân nặng.

```python
mean = np.array([165, 55])
cov = [[100, 0.4], [0.4, 36]]

X = np.random.multivariate_normal(mean, cov, 5000)
```

```python
height, weight = X[:, 0], X[:, 1]

print("Thống kê chiều cao.")
print("Min:", height.min())
print("Max:", height.max())
print("Q1:", np.quantile(height, 0.25))
print("Q2:", np.quantile(height, 0.5))
print("Q3:", np.quantile(height, 0.75))
```

```python
print("Thống kê cân nặng.")
print("Min:", weight.min())
print("Max:", weight.max())
print("Q1:", np.quantile(weight, 0.25))
print("Q2:", np.quantile(weight, 0.5))
print("Q3:", np.quantile(weight, 0.75))
```

```python
plt.boxplot(height)
plt.title("ĐỒ THỊ BOXPLOT THEO HEIGHT")
plt.show()
```

```python
plt.boxplot(weight)
plt.title("ĐỒ THỊ BOXPLOT THEO WEIGHT")
plt.show()
```

```python
print("Đồ thị theo chiều cao")
height = np.round(height, 0)

plt.subplot(2, 2, 1)
plt.hist(height)
plt.title("ĐỒ THỊ TẦN SỐ")

plt.subplot(2, 2, 2)
plt.hist(height, density=True)
plt.title("ĐỒ THỊ TẦN SUẤT")

plt.subplot(2, 2, 3)
x, pmf = np.unique(height, return_counts=True)
pmf = pmf / len(X)

plt.hist(height, density=True)
plt.plot(x, pmf)

plt.subplot(2, 2, 4)
plt.plot(x, pmf)

plt.show()
```

```python
print("Đồ thị theo cân nặng")
weight = np.round(weight, 0)

plt.subplot(2, 2, 1)
plt.hist(weight)
plt.title("ĐỒ THỊ TẦN SỐ")

plt.subplot(2, 2, 2)
plt.hist(weight, density=True)
plt.title("ĐỒ THỊ TẦN SUẤT")

plt.subplot(2, 2, 3)
x, pmf = np.unique(weight, return_counts=True)
pmf = pmf / len(X)

plt.hist(weight, density=True)
plt.plot(x, pmf)

plt.subplot(2, 2, 4)
plt.plot(x, pmf)

plt.show()
```
